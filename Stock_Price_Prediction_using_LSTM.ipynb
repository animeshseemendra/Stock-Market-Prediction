```
import tensorflow as tf
import sklearn as sk
import numpy as np
import pandas as pd
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
df2 = pd.read_csv(io.BytesIO(uploaded['ge.us.txt']))
```

```
df2.head(5)
```

```
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
plt.plot(df2["Open"])
plt.title('GE stock price history')
plt.ylabel('Price (USD)')
plt.xlabel('Days')
plt.legend(['Open'], loc='upper left')
plt.show()
```

```
plt.figure()
plt.plot(df2["High"])
plt.title('GE stock price history')
plt.ylabel('Price (USD)')
plt.xlabel('Days')
plt.legend(['High'], loc='upper left')
plt.show()
```

```
plt.figure()
plt.plot(df2["Low"])
plt.title('GE stock price history')
plt.ylabel('Price (USD)')
plt.xlabel('Days')
plt.legend(['Low'], loc='upper left')
plt.show()
```

```
plt.figure()
plt.plot(df2["Close"])
plt.title('GE stock price history')
plt.ylabel('Price (USD)')
plt.xlabel('Days')
plt.legend(['Close'], loc='upper left')
plt.show()
```

```
plt.figure()
plt.plot(df2["Open"])
plt.plot(df2["High"])
plt.plot(df2["Low"])
plt.plot(df2["Close"])
plt.title('GE stock price history')
plt.ylabel('Price (USD)')
plt.xlabel('Days')
plt.legend(['Open','High','Low','Close'], loc='upper left')
plt.show()
```

```
plt.figure()
plt.plot(df2['Volume'])
plt.xlabel('Days')
plt.ylabel('Volume')
plt.title('GE stock volume history')
plt.show()
```
